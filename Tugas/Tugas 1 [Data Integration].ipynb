```python
### Data integration
import pandas as pd
import numpy as np
```

```python
### Input Data dari EXCEL berdasarkan SHEET
fasilitas = pd.read_excel('tabel1.xls',sheet_name='fasilitas')
korbanjiwa = pd.read_excel('tabel1.xls',sheet_name='korbanjiwa')
rumah = pd.read_excel('tabel1.xls',sheet_name='rumah')
```

```python
## Menjadikan ke Data Frame
fasilitas=pd.DataFrame(fasilitas)
korbanjiwa=pd.DataFrame(korbanjiwa)
rumah=pd.DataFrame(rumah)
```

```python
## Memunculkan data fasilitas
fasilitas
```

```python
## Memunculkan data korbanjiwa
korbanjiwa
```

```python
#Memunculkan Data rumah
rumah
```

```python
#Menggabungkan 3 data fasilitas, rumah dan korbanjiwa berdasarkan kode bencana dan jenis bencana
data=pd.merge(fasilitas,rumah,how='left',on=['Kode Bencana','Jenis Bencana'])
datafinal=pd.merge(data,korbanjiwa,how='left',on=['Kode Bencana','Jenis Bencana'])
datafinal
```

```python
#Menjadikan Kode Bencana sebagai index
datafinal2=datafinal.set_index('Kode Bencana')
datafinal2
```

```python
#Memunculkan datafinal2 hanya 5 data teratas
datafinal2.head()
```

```python
#Memunculkan datafinal2 hanya 5 data belakang
datafinal2.tail()
```

```python
#Menambahkan column Total Kerusakan pada Fasilitas, korbanjiwa dan rumah
datafinal2['Total Fasilitas'] =(datafinal2['Kesehatan']+datafinal2['Peribadatan']+datafinal2['Pendidikan'])
datafinal2['Total Korban Jiwa'] =(datafinal2['Meninggal & Hilang']+datafinal2['Luka-luka']+datafinal2['Menderita & mengungsi'])
datafinal2['Total Rumah'] =(datafinal2['Rusak Berat']+datafinal2['Rusak Sedang']+datafinal2['Rusak Ringan']+datafinal2['Terendam'])
datafinal2
```

```python
# Ordering variable/column
datafinal3 = datafinal2[['Jenis Bencana', 'Kesehatan', 'Peribadatan','Pendidikan','Total Fasilitas',
                         'Rusak Berat','Rusak Sedang','Rusak Ringan','Terendam','Total Rumah',
                        'Meninggal & Hilang','Luka-luka','Menderita & mengungsi','Total Korban Jiwa']]
datafinal3
```
